# Frequent itemset mining

```python
# importamos los paquetes necesarion
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mlxtend.frequent_patterns import apriori

% matplotlib inline
```

```python
basket_starters_before = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\basket_starters_before.csv')
#basket_not_starters = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\basket_not_starters.csv')
#all_base = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\all_base.csv')
```

```python
basket_starters_before = basket_starters_before.set_index('session_fake_id')
basket_starters_before = basket_starters_before.dropna(axis = 0, how = 'any')
basket_starters_before.head()
```

```python
basket_starters_before['1'].count()
```

```python
FIM_s_before = apriori(basket_starters_before, min_support=0.04, use_colnames=True)

FIM_s_before['length'] = FIM_s_before['itemsets'].apply(lambda x: len(x))

FIM_s_before.to_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\FIM_s_before.csv')
```

```python
del basket_starters_before
del FIM_s_before
```
