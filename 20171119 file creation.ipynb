```python
# importamos los paquetes necesarion
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mlxtend.frequent_patterns import apriori

% matplotlib inline 
```

```python
df_sessions = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\session_base.tsv', sep = '\t')
df_sessions = df_sessions.set_index('session_fake_id')
```

# Frequent itemset mining

```python
# cargo los dataset en tres partes y los uno
df1 = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\event_base_1.tsv', sep = '\t')
df2 = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\event_base_2.tsv', sep = '\t')
df3 = pd.read_csv('C:\\Users\\OLX - Gonzalo\\Desktop\\tesis\\event_base_3.tsv', sep = '\t')

df_events = pd.concat([df1, df2, df3])
```

```python
basket_before = (df_events[df_events['before_or_not_start'] ==1]
          .groupby(['session_fake_id', 'event_id'])['session_fake_id']
          .count().unstack().reset_index().fillna(0)
          .set_index('session_fake_id'))
```

```python
basket_before.columns
```

```python
all_base = df_sessions.merge(basket_before, left_index = True, right_index = True, how = 'left')
```

```python
all_base[all_base[2]==0][3].count() / all_base[2].count()
```

```python
frequent_itemsets = apriori(basket_before.sample(n = 1000, replace = False), min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets[ (frequent_itemsets['length'] == 3)]
```

```python
#df_sessions[['starter', 'finisher']].head()

df_events_starters = df_events.set_index('session_fake_id').merge(df_sessions[['starter']], left_index = True, right_index = True, how = 'left')
```

```python
n_events_not_starters = df_events_starters['event_id'][df_events_starters['starter'] == 0].count()
n_events_starters_before = df_events_starters['event_id'][(df_events_starters['starter'] == 1)
                                                       & (df_events_starters['before_or_not_start'] ==1)].count()

df_events_starters[['event_id', 'starter']][df_events_starters['starter'] == 0].groupby(['event_id']).count() / n_events_not_starters 

df_events_starters[['event_id', 'starter']][(df_events_starters['starter'] == 1)
                                                       & (df_events_starters['before_or_not_start'] ==1)].groupby(['event_id']).count() / n_events_starters_before 
```
